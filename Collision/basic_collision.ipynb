```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_circle(x,y,r,c='b'):
    x_c = r*np.cos( np.linspace(-np.pi,np.pi))+x
    y_c = r*np.sin( np.linspace(-np.pi,np.pi))+y
    plt.plot(x_c,y_c,c=c)
```

```python
def generate(N = 4):
    x = np.random.rand(N)*3
    y = np.random.rand(N)*1.6
    r = 0.2 + np.random.rand(N)/7
    c = 0.85 + (np.random.rand(N)-0.5)/7
    a = 0.5 + (np.random.rand(N)-0.5)
    q = 0.9*a+0.1*c
    return x,y,r,q
```

```python
def show(x,y,r):
    plt.scatter(x,y,c='black')
    for x_,y_,r_ in zip(x,y,r):
        plot_circle(x_,y_,r_)
```

```python
x,y,r,q = generate(N = 4)
show(x,y,r)
```
