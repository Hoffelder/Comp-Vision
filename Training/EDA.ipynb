```python
def normalize_sizes(df):
    sizes = []
    for i in tqdm(range(len(df["h"]))):
        if df["h"].iloc[i]>df["w"].iloc[i]:
            p = df["h"].iloc[i]/380
            #print("original size:",df["h"].iloc[i],df["w"].iloc[i])
            #print("normalized:   ",380,int(df["w"].iloc[i]/p))
            sizes.append((380,int(df["w"].iloc[i]/p)))

        else:
            p = df["w"].iloc[i]/380
            #print("original size:",df["h"].iloc[i],df["w"].iloc[i])
            #print("normalized:   ",int(df["h"].iloc[i]/p),380)
            sizes.append((int(df["h"].iloc[i]/p),380))
        
    df["normalized"] = sizes
    
normalize_sizes(df)

```
